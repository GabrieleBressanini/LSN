# Esercizio 04.1

In questo esercizio ci è stato chiesto di modificare il codice di Dinamica Molecolare fornito, in modo di dare la possibilità di utilizzare la configurazione finale della run precedente come configurazione iniziale della nuova simulazione. E'stato inoltre aggiunto un meccanismo che riscala le velocità in modo da ottenere la temperatura voluta.
Queste aggiunte si sono rivelate molto utili nella fase preliminare di equilibrazione della simulazione. Infatti tutte le misure proposte negli esercizi seguenti vanno effettuate quando il sistema è già stato equilibrato e si trova alla temperatura target. Di seguito mostro l'andamento della temperatura (in unità di Lennard Jones) nella fase di equilibrazione, con $T_{target} = 1.1$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]
f1 = np.loadtxt("Esercizio04.1/output_temp.dat")
x1 = np.arange(len(f1))
plt.title("Temperatura su più run")
plt.plot(x1, f1)
plt.ylabel("T")
plt.xlabel("Numero Misure")
plt.axhline(1.1, color='red')
plt.grid(True)
plt.show()
```

Come possiamo vedere, dopo pochi restart del codice il sistema ha raggiunto la temperatura target. Di seguito viene mostrato l'andamento dell'energia cinetica, potenziale e totale per particella in questa fase di equilibrazione, sempre in unità di Lennard Jones. Si noti come il codice riesce a mantenere stabile l'energia totale.

```python
f2 = np.loadtxt("Esercizio04.1/output_ekin.dat")
x2 = np.arange(len(f2))
f3 = np.loadtxt("Esercizio04.1/output_epot.dat")
x3 = np.arange(len(f3))
f4 = np.loadtxt("Esercizio04.1/output_etot.dat")
x4 = np.arange(len(f4))
plt.plot(x2, f2,label="Energia Cinetica")
plt.plot(x3, f3,label="Energia Potenziale")
plt.plot(x4, f4,label="Energia Totale")
plt.ylabel("Energia")
plt.xlabel("Numero Misure")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

Successivamente, come richiesto dagli esercizi 04.2 e 04.3, è stato aggiunto al codice la possibilità di calcolare il valore della pressione e i valori medi, corredati dalle relative incertezze (data blocking), di questa e delle grandezze precedentemente menzionate.
