# 1.Circonferenza

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
x, f = np.loadtxt("Circonferenza/bestcost.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,f,label='costo minore')
plt.title("Costo minore")
plt.xlabel("Temperatura")
plt.ylabel("Costo")
plt.legend()
plt.grid(1)
plt.show()
```

```python
plt.figure(figsize=(8,8))
x, y = np.loadtxt("Circonferenza/bestpath.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,y,'ro')
plt.plot(x,y)
plt.title('Miglior Percorso')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(1)
plt.show()
```
