# Esercizio 10.1

In questo esercizio abbiamo adattato il codice dell'esercitazione precedente in modo tale da risolvere il problema del Travelling Salesman mediante l'algoritmo di Simulated Annealing. Ancora una volta abbiamo studiato il caso in cui le 30 città sono disposte casualmente sulla circonferenza ed il caso in cui queste sono disposte casualmente all'interno di un quadrato. Di seguito viene mostrata la lunghezza del percorso più corto in funzione della temperatura ed infine una rappresentazione grafica del percorso più corto in assoluto trovato dalla simulazione.


# Circonferenza

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]

x, f = np.loadtxt("Esercizio10.1/Circonferenza/bestcost.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,f,label='costo minore')
plt.title("Costo minore")
plt.xlabel("Temperatura")
plt.ylabel("Costo")
plt.legend()
plt.grid(1)
plt.show()
```

```python
plt.figure(figsize=(8,8))
x, y = np.loadtxt("Esercizio10.1/Circonferenza/bestpath.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,y,'ro')
plt.plot(x,y)
plt.title('Miglior Percorso')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(1)
plt.show()
```

Si noti come l'algoritmo è riuscito a trovare il percorso migliore in assoluto.

# Esercizio 10.2

Abbiamo quindi parallelizzato il codice dell'esercizio precedente per effettuare una ricerca parallela del percorso ottimale. Ognuno dei nodi, cinque nel nostro caso, effettua una ricerca indipendente del miglior percorso mediante Simulated Annealing. Alla fine della simulazione i nodi comunicano al nodo "zero" il loro miglior percorso e questo valuta quale sia il migliore in assoluto. Di seguito vengono mostrati la lunghezza del percorso più corto in funzione della temperatura per ognuno dei cinque nodi e successivamente una rappresentazione grafica del percorso ottimale.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x0, f0 = np.loadtxt("Esercizio10.2/bestcost0.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x0,f0,label='Nodo 0')
x1, f1 = np.loadtxt("Esercizio10.2/bestcost1.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x1,f1,label='Nodo 1')
x2, f2 = np.loadtxt("Esercizio10.2/bestcost2.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x2,f2,label='Nodo 2')
x3, f3 = np.loadtxt("Esercizio10.2/bestcost3.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x3,f3,label='Nodo 3')
x4, f4 = np.loadtxt("Esercizio10.2/bestcost4.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x4,f4,label='Nodo 4')
plt.title("Costo minore")
plt.xlabel("Temperatura")
plt.ylabel("Costo")
plt.legend()
plt.grid(1)
plt.show()
```

Come si può vedere, i nodi 1 e 4 hanno trovato il percorso migliore:

```python
plt.figure(figsize=(8,8))
x, y = np.loadtxt("Esercizio10.2/bestpath4.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,y,'ro')
plt.plot(x,y)
plt.title('Miglior Percorso')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(1)
plt.show()
```
