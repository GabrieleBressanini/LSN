In questa soluzione includere anche i 4 grafici a blocchi, di una run un particolare, delle 4 grandezze.
Nel senso: fare vedere tutta la progressione dell'errore con i 20 blocchi (dato che nell'immagine richiesta io faccio uso solamente dell'ultimo valor medio e dell'errore ad esso associato...).
Si noti che la misura di $M$ Ã¨ l'unica fatta con $h\neq0$, le altre sono tutte con $h=0$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
x, f, error = np.loadtxt("output.ene.10", usecols=(0,2,3), delimiter='   ', unpack='true')
xG, fG, errorG = np.loadtxt("outputG.ene.10", usecols=(0,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, color="cyan", label= "Metropolis")
plt.errorbar(xG,fG,yerr=errorG, color="orange", label= "Gibbs")
plt.xlabel('blocks')
plt.ylabel('Internal Energy U')
plt.grid(True)
plt.legend()
plt.show()
```

```python
block, I, error = np.loadtxt("output.heat.0",usecols=(0,2,3), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Calore Specifico")
plt.xlabel("Numero di blocchi")
plt.ylabel("C")
plt.grid(True)
plt.show()
```

```python
block, I, error = np.loadtxt("output.chi.0",usecols=(0,2,3), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("SuscettivitÃ  Magnetica")
plt.xlabel("Numero di blocchi")
plt.ylabel("X")
plt.grid(True)
plt.show()
```
