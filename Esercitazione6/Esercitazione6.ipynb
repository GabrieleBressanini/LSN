In questo esercizio ci è stato inizialmente chiesto di completare il codice fornito, in modo da simulare un modello di Ising 1D. In particolare abbiamo implementato due algoritmi di sampling: il Metropolis e il Gibbs. E' stata inoltre introdotta l'opzione di utilizzare la configurazione finale della run precedente come configurazione iniziale della nuova simulazione. Successivamente abbiamo aggiunto la possibilità di misurare la capacità termica, la magnetizzazione e la suscettibilità magnetica. Le unità di misura utilizzate sono tali per cui $\mu_B=1$ e $k_B=1$.

Nei grafici seguenti mostro l'andamento, per un sistema di $N=50$ spin aventi $J=1$, delle funzioni $U(N,T)$, $C(N,T)$, $\chi(N,T)$, $M(N,T,h=0.02)$ con le relative incertezze (data blocking), per $T\in [0.5,2]$. Queste curve vengono confrontate con quelle teoriche, calcolate analiticamente.
Si noti che uno step Monte Carlo consiste nel provare a "girare" tutti gli spin uno alla volta e che, prima di effettuare ogni tipo di misura, il sistema è stato equilibrato.

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e, label="Exact")

x, f, error = np.loadtxt("Esercizio06.1/EnergyM.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
xG, fG, errorG = np.loadtxt("Esercizio06.1/EnergyG.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error, color="cyan", label= "Metropolis")
plt.errorbar(xG,fG,yerr=errorG, color="orange", label= "Gibbs")
plt.xlabel('T')
plt.ylabel('U')
plt.title('Energia Interna')
plt.grid(True)
plt.legend()
plt.show()
```

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label="Exact")

x, f, error = np.loadtxt("Esercizio06.1/HeatM.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
xG, fG, errorG = np.loadtxt("Esercizio06.1/HeatG.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error, color="cyan", label= "Metropolis")
plt.errorbar(xG,fG,yerr=errorG, color="orange", label= "Gibbs")
plt.xlabel('T')
plt.ylabel('C')
plt.title("Calore Specifico")
plt.grid(True)
plt.legend()
plt.show()

```

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M, label="Exact")

x, f, error = np.loadtxt("Esercizio06.1/MagM.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
xG, fG, errorG = np.loadtxt("Esercizio06.1/MagG.txt", usecols=(0,3,4), delimiter='   ', unpack='true')
plt.errorbar(x,f,yerr=error, color="cyan", label= "Metropolis")
plt.errorbar(xG,fG,yerr=errorG, color="orange", label= "Gibbs")
plt.xlabel('T')
plt.ylabel('Magnetizzazione')
plt.title("M")
plt.grid(True)
plt.legend()
plt.show()
```
