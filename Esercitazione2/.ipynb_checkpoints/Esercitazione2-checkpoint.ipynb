# Esercizio 02.1

In questo esercizio abbiamo calcolato mediante metodi Monte Carlo l'integrale

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

in due modi: utilizzando sampling uniforme in [0,1) e utilizzando successivamente l'importance sampling. Nel secondo caso abbiamo utilizzato, come densità di probabilità con la quale estrarre le variabili casuali, la funzione (già normalizzata nell'intervallo di integrazione) $p(x) = 2(1-x)$. Per ottenere variabili casuali aventi tale distribuzione si utilizza il metodo della funzione cumulativa inversa. Si trova quindi che $1-\sqrt{1-y}$ , dove $y$ è distribuita uniformemente in [0,1), ha la distribuzione corretta. Il nuovo integrando è quello precedente fratto $p(x)$.
Mostriamo di seguito il valore dell'integrale e il relativo errore al variare del numero di blocchi. Ci aspettiamo che l'integrale calcolato mediante importance sampling abbia un errore inferiore rispetto al primo.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]
block, I, error = np.loadtxt("Esercizio02.1/SamplingUniforme.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Valutazione dell'integrale mediante Sampling Uniforme")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Valore Integrale")
plt.axhline(1,color='r')
plt.grid(1)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, I, error = np.loadtxt("Esercizio02.1/ImportanceSampling.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Valutazione dell'integrale mediante Importance Sampling")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Valore Integrale")
plt.axhline(1,color='r')
plt.grid(1)
plt.show()
```

# Esercizio 02.2

In questo esercizio abbiamo simulato un Random Walk 3D discreto su un reticolo cubico e successivamente un Random Walk sul continuo. In entrambi i casi abbiamo ripetuto la simulazione molte volte, ogni volta riposizionando il walker nell'origine degli assi. Mostro, in entrambi i casi, l'andamento di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ con le relative incertezze, al variare dello step $i\in [0,100]$. Gli andamenti trovati sono quelli aspettati, ovvero del tipo $f(N)=k\sqrt{N}$.

```python
block, I, error = np.loadtxt("Esercizio02.2/RandomWalkDiscreto.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Random Walk Discreto")
plt.xlabel("Step")
plt.ylabel("Distanza dall'origine")
plt.grid(1)
plt.show()
```

```python
block, I, error = np.loadtxt("Esercizio02.2/RandomWalkContinuo.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Random Walk Continuo")
plt.xlabel("Step")
plt.ylabel("Distanza dall'origine")
plt.grid(1)
plt.show()
```
