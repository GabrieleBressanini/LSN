# Esercizio 01.1

In questo esercizio ci è stato richiesto di calcolare l'integrale $\langle r \rangle = \int_0^1 r dr = 1/2$ e la sua varianza $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ con metodi Monte Carlo, per testare il generatore di numeri casuali fornito. Viene mostrato l'andamento di $<r>$ e della varianza $\sigma^2$, corredati dai relativi errori, in funzione del numero di blocchi. Come si può notare, i risultati ottenuti sono compatibili con i valori veri entro l'errore del calcolo.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,9))
block, I, error = np.loadtxt("Esercizio01.1/results.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("<r>")
plt.xlabel("Numero di blocchi")
plt.ylabel("<r>")
plt.axhline(y=0.5)
plt.grid(1)
plt.show()

plt.figure(figsize=(16,9))
block2, I2, error2 = np.loadtxt("Esercizio01.1/results2.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block2,I2,yerr=error2)
plt.title(r"$\sigma^2$")
plt.xlabel("Numero di blocchi")
plt.ylabel(r"$\sigma^2$")
plt.axhline(y=1./12.)
plt.grid(1)
plt.show()
```

Infine abbiamo suddiviso l'intervallo [0,1] in 100 intervalli uguali e abbiamo implementato il test del $\chi^2$, usando ogni volta $10^4$ numeri distribuiti uniformemente. Ci aspettiamo dunque che  $\chi^2 \simeq100$ . Come possiamo vedere dal plot successivo, i valori di $\chi^2_i$ sono distrubuiti attorno a tale valore.

```python
plt.figure(figsize=(16,9))
f = np.loadtxt("Esercizio01.1/results3.txt")
y_chi = f[:,1]
x_chi = np.arange(len(y_chi))
plt.scatter(x_chi,y_chi)
plt.title(r"$\chi^2$")
plt.xlabel(r"$\chi^2_i$")
plt.ylabel(r"$\chi^2$")
plt.grid(1)
plt.show()
```

# Esercizio 01.3

In questo esercizio abbiamo dovuto simulare l'esperimento di Buffon: un ago di lunghezza $L$ viene lasciato cadere casualmente su un piano orizzontale decorato da linee parallele distanti $d$, con $d > L$. La probabilitù $P$ che l'ago intersechi una di queste è $P = 2L/\pi d$. Segue quindi che
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
Di seguito viene mostrata la stima di $\pi$ e del suo errore, in funzione del numero di blocchi $N$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, I, error = np.loadtxt("Esercizio01.3/results.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title(r"$\pi$")
plt.xlabel("Numero di blocchi")
plt.ylabel(r"stima $\pi$")
plt.grid(1)
plt.axhline(np.pi,color='red')
plt.show()
```
