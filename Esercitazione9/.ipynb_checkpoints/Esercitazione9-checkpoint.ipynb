# Esercizio 09.1

In questa esercitazione abbiamo implementato un algoritmo genetico per risolvere il problema del travelling salesman. In particolare sono state esplorate due diverse configurazioni del problema: una in cui le città sono disposte casualmente sulla circonferenza goniometrica, ed una in cui le città sono disposte casualmente all'interno di un quadrato di lato 1. In entrambi i casi le città sono 30 e il numero di individui ad ogni generazione 900.
Di seguito vengono mostrati gli andamenti del costo minore e della media dei costi della metà migliore della popolazione al variare della generazione.
Infine verrà mostrato graficamente il percorso migliore.

# 1.Circonferenza

In questo conosciamo già a priori il percorso migliore, quindi possiamo verificare la bontà dell'algoritmo.
Dato il numero delle città ci aspettiamo che il miglior percorso abbia una lunghezza (costo) "vicina" a $2\pi \simeq 6.28$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
x, f = np.loadtxt("Circonferenza/bestcost.txt",usecols=(0,1), delimiter="   ", unpack='true')
x1, f1 = np.loadtxt("Circonferenza/bestavg.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,f,label='costo minore')
plt.plot(x1,f1,label='media costo')
plt.title("Costo minore e media del costo della metà migliore della popolazione")
plt.xlabel("Generazione")
plt.ylabel("Costo")
plt.legend()
plt.grid(1)
plt.show()
```

```python
plt.figure(figsize=(8,8))
x, y = np.loadtxt("Circonferenza/bestpath.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,y,'ro')
plt.plot(x,y)
plt.title('Miglior Percorso')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(1)
plt.show()
```

# 3.Curiosità

Il codice riesce a trovare il minimo anche in assenza del crossover. In questo caso, tuttavia, sono necessarie più generazioni ed è stata aumentata la probabilità di avere una mutazione. Ad ogni generazione viene scelta e duplicata la metà migliore della popolazione. Di seguito si mostrano i risultati per le città disposte sulla circonferenza.

```python
plt.figure(figsize=(10,6))
x, f = np.loadtxt("Curiosita/bestcost.txt",usecols=(0,1), delimiter="   ", unpack='true')
plt.plot(x,f,label='costo minore')
plt.title("Costo minore")
plt.xlabel("Generazione")
plt.ylabel("Costo")
plt.legend()
plt.grid(1)
plt.show()
```
