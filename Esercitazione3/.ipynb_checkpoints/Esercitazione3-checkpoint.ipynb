# Esercizio 03.1

In questo esercizio abbiamo dovuto valutare il prezzo delle Call e delle Put options europee, utilizzando la teoria di Black e Scholes. Questa assume che l'evoluzione temporale del prezzo di questi asset sia stocastico ed esibisca un moto browniano geometrico, con costante di interesse a rischio zero $r$ e volatilit√† $\sigma$. Per il calcolo del prezzo abbiamo seguito due metodi differenti: inizialmente abbiamo fatto un sampling diretto del prezzo finale, successivamente abbiamo diviso l'intervallo temporale in sotto intervalli e abbiamo seguito l'evoluzione temporale (discreta) del prezzo al variare del tempo. I risultati ottenuti sono compatibili, entro l'errore, con la soluzione analitica di Black e Scholes.

Di seguito mostro l'andamento dei prezzi delle opzioni e le relative incertezze in funzione del numero di blocchi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]
block, I, error = np.loadtxt("Esercizio03.1/PutDirect.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Prezzo Put - Metodo Diretto")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Prezzo Put")
plt.axhline(y=5.4595325819072364)
plt.grid(1)
plt.show()
```

```python
block, I, error = np.loadtxt("Esercizio03.1/CallDirect.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Prezzo Call - Metodo Diretto")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Prezzo Call")
plt.axhline(y=14.975790778311286)
plt.grid(1)
plt.show()
```

```python
block, I, error = np.loadtxt("Esercizio03.1/PutDiscretized.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Prezzo Put - Metodo Discreto")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Prezzo Put")
plt.axhline(y=5.4595325819072364)
plt.grid(1)
plt.show()
```

```python
block, I, error = np.loadtxt("Esercizio03.1/CallDiscretized.txt",usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,I,yerr=error)
plt.title("Prezzo Call - Metodo Discreto")
plt.xlabel("Numero di Blocchi")
plt.ylabel("Prezzo Call")
plt.axhline(y=14.975790778311286)
plt.grid(1)
plt.show()
```
