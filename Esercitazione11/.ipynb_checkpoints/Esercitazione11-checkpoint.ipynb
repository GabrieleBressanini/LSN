# Esercizio 11.01

In questo esercizio abbiamo effettuato il fit di una funzione lineare modificando il Network Neurale fornito. Quest'ultimo ha un solo neurone, e quindi due iperparametri da ottimizzare. In particolare abbiamo aggiunto del rumore ai dati di training e abbiamo aumentato il numero di epoche.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.2 # noise standard deviation
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 100 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=100,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

# Esercizio 11.02

In questo esercizio abbiamo implementato un Network Neurale differente per effettuare il fit del polinomio 

$$
f(x)=4-3x-2x^2+3x^3
$$

per $x \in [-1,1]$. Rispetto all'esercizio precedente abbiamo aumentato i dati di training e di validation ed abbiamo utilizzato, come ottimizzatore, l'Adam. Il network consiste di cinque layer densi aventi 4 neuroni ciascuno. La funzione di attivazione utilizzata è la elu. Il layer di output, ovviamente, ha un solo neurone e nessuna funzione di attivazione.

```python
#parameters 
a = 4.
b = -3.
c = -2.
d = 3.

def poly(x):
    return a + b * x + c*x**2 + d*x**3
```

```python
# generate training inputs
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target =  poly(x_valid)# ideal (target) function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(poly(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(poly(x_valid), sigma)

# plot training and target dataset
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.show()
```

```python
model = tf.keras.Sequential()

model.add(Dense(4, input_shape=(1,), activation='elu'))
model.add(Dense(4, activation='elu'))
model.add(Dense(4, activation='elu'))
model.add(Dense(4, activation='elu'))
model.add(Dense(4, activation='elu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mse'])

model.summary()
```

```python
history = model.fit(x=x_train,
                    y=y_train,
                    batch_size=8,
                    epochs=30,
                    shuffle=True, # a good idea is to shuffle input before at each epoch
                    validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
 
```

# Esercizio 11.03

In questo esercizio abbiamo dovuto effettuare il fit della seguente funzione di due varabili:

$$f(x,y) = \sin(x^2+y^2)$$ 

per $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$. Per farlo abbiamo modificato il network dell'esercizio precedente, innanzitutto per far si che accettasse dati bidimensionali.

```python
def f(x):
    return np.sin(x[:,0]**2.+x[:,1]**2.)
```

```python
# generate training inputs
np.random.seed(0)

n_data = 10000
n_validation = int(n_data*0.1)

x_train = np.zeros((n_data,2))
x_valid = np.zeros((n_validation,2))

x_train[:,0] = np.random.uniform(-1.5, 1.5, n_data)    
x_train[:,1] = np.random.uniform(-1.5,1.5,n_data)

x_valid[:,0] = np.random.uniform(-1.5, 1.5, n_validation)    
x_valid[:,1] = np.random.uniform(-1.5, 1.5, n_validation)

y_target = f(x_valid) # ideal (target) function

sigma = 0.0 # noise standard deviation

y_train = np.random.normal(f(x_train), sigma) # actual measures from which we want to guess regression parameters

y_valid = np.random.normal(f(x_valid), sigma)
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_target

fig = plt.figure()

ax = Axes3D(fig)

ax.scatter(X, Y, Z, marker="+",color='r', label="Validation Set")

X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_valid

ax.plot(X, Y, Z, marker='+',label="Th", linestyle="None")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 60)
plt.legend()
plt.show()
```

```python
model = tf.keras.Sequential()

model.add(Dense(4, input_shape=(2,), activation='selu'))
model.add(Dense(4, activation='selu'))
model.add(Dense(4, activation='selu'))
model.add(Dense(4, activation='selu'))
model.add(Dense(4, activation='selu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mse'])

model.summary()
```

```python
history = model.fit(x = x_train,
                    y = y_train, 
                    batch_size=8,
                    epochs=30,
                    shuffle=True,
                    validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
n_predict = 1000
x_predicted = np.zeros((n_predict, 2))
x_predicted[:,0] = np.random.uniform(-1.5, 1.5, n_predict)
x_predicted[:,1] = np.random.uniform(-1.5, 1.5, n_predict)
y_predicted = model.predict(x_predicted)



X = x_predicted[:,0]
Y = x_predicted[:,1]
Z = y_predicted

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X, Y, Z, marker='o', color='b', label='Predicted')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# # # 
X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_target

ax.scatter(X, Y, Z, marker='o', color='green', label='Target')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(40, 60)

plt.legend(loc='best')
plt.show()
```
