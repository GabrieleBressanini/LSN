# Esercizio 05.1

In questo esercizio abbiamo utilizzato l'algoritmo di Metropolis per fare sampling della distribuzione di probabilità associata agli autostati $\psi_{1,0,0}$ (1s) e $\psi_{2,1,0}$ (2p) dell'atomo di idrogeno al fine di calcolare $\langle r \rangle$. Tutte le lunghezze sono espresse in unità del raggio di Bohr $a_0$. Come al solito valori medi e relative incertezze sono state calcolate in funzione del numero di blocchi. Come è possibile valori, i calcoli Monte Carlo sono compatibili con i valori teorici noti: $$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
Abbiamo utilizzato, come probabilità di transizione, la distribuzione uniforme e quella gaussiana, dove lo step è stato scelto in modo tale da assicurare un'accetazione del 50% circa. Per lo stato 1s il punto di partenza è stato posto nell'origine degli assi, mentre per il 2p si è scelto il punto (0,0,1).

Di seguito uno scatterplot per visualizzare in 3D i punti estratti secondo le due distribuzioni.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =  np.loadtxt("Esercizio05.1/graph1s.txt", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.title("1s")
plt.show()
```

# Distribuzione Uniforme

```python
block, r0, error = np.loadtxt("Esercizio05.1/1s.txt", usecols=(0,1,2), delimiter="   ", unpack='true')
plt.errorbar(block,r0,yerr=error)
plt.axhline(y=1.5,color='red')
plt.title(r"$\langle r \rangle$ (1s)")
plt.xlabel('Numero Blocchi')
plt.ylabel(r'$\langle r \rangle$')
plt.grid(1)
plt.show()
```
